### Pertanyaan 1: Seberapa besar perbedaan penggunaan sepeda pada musim panas, musim gugur, musim dingin, dan musim semi dalam rentang waktu 2011-2012?

Dari grafik tersebut, terlihat bahwa jumlah sewa sepeda terlihat paling banyak pada musim gugur, kemudian musim panas, lalu disusul dengan musim dingin, dan terakhir musim semi. Grafik ini memberikan informasi penting bagi pengelola layanan sepeda untuk mengatur stok sepeda sesuai dengan musim.
Mengingat jumlah pengguna paling tinggi ada pada musim gugur, diikuti oleh musim panas dan musim dingin, dan terendah pada musim semi, pengelola dapat menyesuaikan strategi pemasaran dan pengelolaan inventaris sepeda berdasarkan data ini. Strategi promosi mungkin perlu ditingkatkan pada musim semi untuk menarik lebih banyak pengguna, sementara strategi pemeliharaan dan pengelolaan stok bisa difokuskan lebih pada musim gugur.

# Pertanyaan 3 : Bagaimana pola penggunaan sepeda berubah selama akhir pekan (hari libur) dibandingkan dengan hari-hari kerja?

```python
week_mapping = {0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'}
df_day['weekday'] = df_day['weekday'].map(week_mapping)

fig = px.bar(df_day, x="weekday", y="cnt", title="Grafik Hari dengan Jumlah Pengguna")
fig.update_xaxes(title="Hari")
fig.update_yaxes(title="Jumlah Pengguna")
fig.show()

# Mengonversi tipe data kolom 'dteday' menjadi tipe data datetime
df_day['dteday'] = pd.to_datetime(df_day['dteday'])

# Ekstrak hari dari kolom 'dteday' (0: Senin, 1: Selasa, ..., 6: Minggu)
df_day['day_of_week'] = df_day['dteday'].dt.dayofweek

# Mengidentifikasi apakah hari adalah akhir pekan (0: Tidak, 1: Ya)
df_day['is_weekend'] = df_day['day_of_week'].isin([5, 6]).astype(int)

# Menghitung rata-rata jumlah sewa sepeda berdasarkan akhir pekan dan hari kerja
df_weekend = df_day.groupby('is_weekend')['cnt'].mean().reset_index()

# Visualisasi rata-rata penggunaan sepeda pada akhir pekan dan hari kerja menggunakan bar plot
fig = px.bar(df_weekend, x='is_weekend', y='cnt', color='is_weekend',
             labels={'is_weekend': 'Hari Libur (0: Tidak, 1: Ya)', 'cnt': 'Rata-rata Jumlah Sewa Sepeda'},
             title='Rata-rata Penggunaan Sepeda pada Akhir Pekan vs. Hari Kerja',
             text='cnt')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(xaxis_title='Hari Libur (0: Tidak, 1: Ya)', yaxis_title='Rata-rata Jumlah Sewa Sepeda')
fig.update_xaxes(tickvals=[0, 1], ticktext=['Hari Kerja', 'Akhir Pekan'])
fig.show()
```

Analisis dari visualisasi ini menyoroti perbedaan dalam rata-rata penggunaan sepeda antara hari kerja dan akhir pekan. Dengan menggunakan data waktu, kita dapat mengidentifikasi hari-hari dalam seminggu dan memisahkan mereka menjadi hari kerja dan akhir pekan. Hasilnya menunjukkan bahwa rata-rata jumlah sewa sepeda cenderung lebih tinggi pada akhir pekan daripada pada hari kerja. Hal ini mungkin disebabkan oleh banyak faktor, seperti waktu luang yang lebih besar pada akhir pekan atau aktivitas rekreasi yang lebih banyak dilakukan pada hari libur. Visualisasi menggunakan bar plot dengan jelas memperlihatkan perbedaan ini, dengan rata-rata jumlah sewa sepeda yang lebih tinggi pada hari-hari libur (nilai 1) dibandingkan dengan hari-hari kerja (nilai 0). Analisis ini memberikan wawasan yang berguna bagi penyedia layanan sewa sepeda untuk mengatur strategi operasional dan pemasaran mereka, seperti menyesuaikan persediaan sepeda atau menargetkan promosi khusus untuk meningkatkan penggunaan sepeda pada akhir pekan.
